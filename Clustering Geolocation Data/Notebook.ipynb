```python
class_predictions = np.load('Data/sample_clusters.npy')
```

```python
plt.scatter(x=X_blob[:,0], y=X_blob[:,1], c=class_predictions, alpha=0.2)
```

Let us use the **silhouette score**, which is the mean silhouette coefficient of all samples. The **silhouette coefficient** is calculated using the mean intra-cluster distance ($a$) and the mean nearest-cluster distance ($b$) for each sample. The silhouette coefficient for a sample is $\frac{b-a}{\max(a,b)}$. To clarify, $b$ is the distance between a sample and the nearest cluster that the sample is not a part of.

- $b$ = mean distance to the points in the nearest cluster

- $a$ = mean intra-cluster distance to all the points

The range of Silhouette score is $[-1, 1]$. Its analysis is as follows:

$+1$ Score − Near $+1$ Silhouette score indicates that the sample is far away from its neighboring cluster.

$0$ Score − indicates that the sample is on or very close to the decision boundary separating two neighboring clusters.

$-1$ Score −  indicates that the samples have been assigned to the wrong clusters.

```python
silhouette_score(X_blob, class_predictions)
```

# K-Means Clustering

```python
X_blob, _ = make_blobs(n_samples=1000, centers=50,
                          cluster_std=1, random_state=4)
```

```python
sns.scatterplot(x=X_blob[:,0], y=X_blob[:,1], alpha=0.3)
```

```python
# look at 20 iterations of the K-Means algorithm
k = 5
data_dict = defaultdict(dict)
for x in range(1,21):
    model = KMeans(n_clusters=k, random_state=17, n_init=1, max_iter=x)
    data_dict[x]['class_predictions'] = model.fit_predict(X_blob)
    data_dict[x]['centroids'] = model.cluster_centers_
```

```python
def f(x):
    plt.scatter(x=X_blob[:,0], 
                y=X_blob[:,1], 
                c=data_dict[x]['class_predictions'], 
                alpha=0.2)
    plt.scatter(x=data_dict[x]['centroids'][:, 0], 
                y=data_dict[x]['centroids'][:, 1], 
                c='#000000', 
                marker='v',
                s=100)
    plt.title('K-Means Neat Visual')
```

```python
interactive(f, x=(1, 20))
```

```python
X = np.array(data[['LON', 'LAT']], dtype='float64')
```

```python
def create_map(k):
    model = KMeans(n_clusters=k, random_state=17)
    class_predictions = model.fit_predict(X)
    data['CLASS'] = class_predictions
    cols = random.sample(list(matplotlib.colors.cnames.values()), k)
    
    m = folium.Map(location=[np.average(data['LAT']), np.average(data['LON'])],
              tiles='Stamen Terrain', zoom_start=9)
    
    for _, row in data.iterrows():
        folium.CircleMarker(
            location=[row['LAT'], row['LON']],
            radius=5,
            popup=re.sub(r'[^a-zA-Z ]+','',row['NAME']),
            color=cols[row['CLASS']],
            fill=True,
            fill_colour=cols[row['CLASS']]
        ).add_to(m)
    
    return m, class_predictions
```

```python
m, class_predictions = create_map(70)
print(f'Silhouette Score: {silhouette_score(X, class_predictions)}')
m
```

```python
m.save('kmeans_map.html')
```

```python
# depends on use-case
best_silhouette, best_k = -1, 0

for k in tqdm(range(2,100)):
    model = KMeans(n_clusters=k, random_state=1)
    class_predictions = model.fit_predict(X)
    
    curr_silhouette = silhouette_score(X, class_predictions)
    if curr_silhouette > best_silhouette:
        best_k = k
        best_silhouette = curr_silhouette

print(f'Best K: {best_k}')
print(f'Best Silhouette: {best_silhouette}')
```
